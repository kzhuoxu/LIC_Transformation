```python
# load Parks_Properties_20241031.csv in geopandas

import geopandas as gpd

data = gpd.read_file('Parks_Properties_20241031.csv')

data.head()
```

```python
# the multipolygon is string type, convert it to a multipolygon type
from shapely.wkt import loads
data['geom'] = data['multipolygon'].apply(loads)

# find the centroid of each polygon. note, some polygons are of type Series
data['centroid'] = data['geom'].apply(lambda x: x.centroid)
```

```python
#  convert data[['centroid', 'EAPPLY']] to new geodataframe
gdf = gpd.GeoDataFrame(data[['centroid', 'EAPPLY']], geometry='centroid')

# rename EAPPLY to park_name
gdf.rename(columns={'EAPPLY':'park_name'}, inplace=True)

# save the geodataframe to a json file

gdf.to_file('parks.json', driver='GeoJSON')


```
